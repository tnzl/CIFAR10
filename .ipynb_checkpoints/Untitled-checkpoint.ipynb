```python
import pickle
import numpy as np
import tensorflow.keras as keras
from sklearn.preprocessing import OneHotEncoder, normalize
import cv2
from prepdata import Data_processor
import cv2
import matplotlib.pyplot as plt
```

```python
def load_cfar10_batch(cifar10_dataset_folder_path, batch_id):
    with open(cifar10_dataset_folder_path + '/data_batch_' + str(batch_id), mode='rb') as file:
        # note the encoding type is 'latin1'
        batch = pickle.load(file, encoding='latin1')
        
    features = batch['data'].reshape((len(batch['data']), 3, 32, 32)).transpose(0, 2, 3, 1)
    labels = batch['labels']
    
    enc=OneHotEncoder(categories='auto')
    labels=enc.fit_transform(np.array(labels).reshape(-1,1))
    
    return features, labels
```

```python
IMG_SIZE  = 32
xx=[]
yy=[]

for i in range(1,6):
    x,y=load_cfar10_batch('./cifar-10-batches-py',i)
    xx+=[x]
    yy+=[y]
x = np.array(xx)
#.reshape(5,10000,3,32,32)
y = np.array(yy)

x.shape,y.shape
```

```python
IMG_SIZE  = 32
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)

# Create the base model from the pre-trained model MobileNet V2
base_model = keras.applications.ResNet50(input_shape=IMG_SHAPE,
                                               include_top=False,
                                               weights='imagenet')
```

```python
base_model.trainable=False
```

```python
model = keras.Sequential([
    base_model,
    keras.layers.Flatten(),
    keras.layers.Dense(128),
    keras.layers.Dense(10)
])
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=3e-2),
              loss=keras.losses.categorical_crossentropy,
              metrics=['accuracy'],)
```

```python
for epoch in range(3):
    for i in range(5):
        model.fit(x[i],y[i],epochs=1)
    
```
