```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
import CONSTANTS as c
```

```python
model_name = 'Basic'
attempt = 1
```

```python
record = None
with open(c.ROOT[:-4] + 'logs/' + record_name(model_name, attempt) + '.pkl', 'rb') as f:
    record = pickle.load(f)
    
%matplotlib inline
n = len(record.keys())
from math import ceil
rows = ceil(n/4)
fig = plt.figure(figsize=(20, 5*rows))
i = 0
for key in record.keys():
    i += 1
    fig.add_subplot(rows, 4, i)
    plt.plot(record[key])
    plt.title(key)
```

```python
def scale(arr, x_max):
    n = len(arr)
    d = x_max//n
    l = [arr[0]] * d
    for i in range(1, n):
        l += list(np.linspace(arr[i-1], arr[i], num=d))
    return l 
                
        
fig = plt.figure(figsize=(10, 5))
fig.add_subplot(1, 2, 1)
plt.plot(record['loss'], )
plt.plot(scale(record['val_loss'], len(record['loss'])))
plt.title('loss')

fig.add_subplot(1, 2, 2)
plt.plot(record['accuracy'])
plt.plot(scale(record['val_accuracy'], len(record['accuracy'])))
plt.title('accuracy')
```
